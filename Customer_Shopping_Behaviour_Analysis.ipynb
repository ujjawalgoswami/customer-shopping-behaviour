```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include = 'all')
#include = 'all' -> gives summery of statistical columns for numerical and categorical columns
```

```python
df.isnull().sum()
```

```python
#mean is not used to fill review rating because it has outliers and median is used because it has no outliers.
df['Review Rating'] = df.groupby('Category') ['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#make all the column names in smallcase and also remove spaces and add underscore
#make all columns smallcase
df.columns = df.columns.str.lower()
#add underscore in place of spaces
df.columns = df.columns.str.replace(' ','_')
#rename purchase amount(USD)
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
    
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
#here we can check that both columns have same information(if true then we remove one of them)
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#remove promo_code_used column
df = df.drop('promo_code_used',axis = 1)
```
